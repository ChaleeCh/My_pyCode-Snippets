```python
# COMBINING DATAFRAMES USING CONCAT() AND MERGE() 

# create dummy data
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
 

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                     'B': ['B4', 'B5', 'B6', 'B7'],
                     'C': ['C4', 'C5', 'C6', 'C7'],
                     'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])
 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                     'B': ['B8', 'B9', 'B10', 'B11'],
                     'C': ['C8', 'C9', 'C10', 'C11'],
                     'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])
```

```python
# combine dataframes row-wise
result = pd.concat([df1, df2, df3])
result
```

```python
# label the dataframes to know which data came from where

result = pd.concat([df1, df2, df3], keys=['x', 'y', 'z'])
result
```

```python
# get second dataframe
result.loc['y']
```

```python
# using the merge() function

# create dummy data
df_a = pd.DataFrame({
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']})

df_b = pd.DataFrame({
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']})

df_c = pd.DataFrame({
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]})
```

```python
# merge df_a and df_c based on the subject_id

pd.merge(df_a, df_c, on='subject_id')
```

```python
# outer join used to produce a set of all records in table A and B where available
# if there are no matches, the missing side will contain null values

pd.merge(df_a, df_b, on='subject_id', how='outer')
```

```python
# DATA VISUALIZATION WITH SEABORN
# Loading the dataset

filePath = 'C:/Datasets/'
fileName = 'vgsales.csv'
Data = filePath + fileName
Data_vgSales = pd.read_csv(Data)
Data_vgSales.head()
```

```python
# creating basic plots
# first we create a line plot using relplot of National sales against EU sales

sns.lineplot(x="Year", y="EU_Sales",data=data_vgSales[:30]);
```

```python
# to create a bar plot with seaborn, only need to change the plot type 

sns.barplot(x="Year", y="EU_Sales",data=data_vgSales[:6]);
```

```python
# let's plot histograms to compare the national sales with the EU sales 

sns.distplot(data_vgSales['NA_Sales'])
sns.displot(data_vgSales['EU_Sales'])
```

```python
# box plot

sns.boxplot(data_vgSales['JP_Sales'], orient='horizontal')
```

```python
# violin plot

sns.violinplot(data_vgSales['JP_Sales'], orient='vertical', color='yellow')
```

```python
# we can add another dimension to the plot by coloring the points according to a third variable 
# this is called hue semantic in seaborn

sns.relplot(x="NA_Sales", y="EU_Sales", hue="JP_Sales",data=data_vgSales[:200]);
```

```python
# subplots for each of the category of Outlet_Size
sns.relplot(x="NA_Sales", y="EU_Sales",hue='JP_Sales',style='Other_Sales',col='Other_Sales',data=data_vgSales[:2]);
```
